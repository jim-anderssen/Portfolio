```python
from geopy.distance import distance
from geopy.point import Point
import pandas as pd
import json
```

```python

def assign_circle(name,latitude,longitude):
    y = 63.658708
    x = 22.539284
    center = Point(latitude,longitude)
    radius = 300
    num_points = 36 # every 10 degrees
    locations = {}
    circle_coords = []
    
    
    for angle in range(0,360,int(360/num_points)):
        destination = distance(meters=radius).destination(point=center, bearing=angle)
        circle_coords.append((destination.latitude, destination.longitude))
    circle_coords.append(circle_coords[0])
    return circle_coords
    
```

```python
badstränder = pd.read_csv('../Kallans badstränder.csv',sep=';')
```

```python
geometry_by_name = {}

for _,row in badstränder.iterrows():
    coords = assign_circle(row['Toiminta'].lstrip(),row['Pohjoinen koordinaatti'],row['Itäinen koordinaatti'])
    
    geometry_by_name[row['Toiminta'].lstrip()] = {
        'type':'Polygon',
         'coordinates':[coords]
    }

    features.append(feature)
```
